### Request hourly data
Remember that not all variables are available at hourly time steps. For instance, if you pass"TEMP2MMAX" to the code below it will return all NaN since the mesonet does not expose min and max temperature at hourly time stemps, only daily (min and max of the day).

```python
station = "Manhattan"
start_date = "2025-06-01 01:00:00"
end_date = "2025-06-10 00:00:00"
interval = "hour"
variables = ["PRECIP","TEMP2MAVG"]


df = ms.request_data(station,start_date,end_date,interval,variables)
df.head(5)
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(6, 4))
charts.plot_temperature(axes[0], df, ["TEMP2MAVG"])
charts.plot_precip(axes[1], df, "PRECIP")
plt.show()
```

### Request daily data

```python
station = "Manhattan"
start_date = "2025-01-01"
end_date = "2026-01-01"
interval = "day"
variables = ["TEMP2MAVG","TEMP2MMAX","TEMP2MMIN","SR"]


df = ms.request_data(station,start_date,end_date,interval,variables)
df.head(5)
```

```python
df = ms.request_data('Manhattan',"2025-07-01","2025-08-01",'day',['SRAVG'])
df.head(3)
```

```python
df = ms.request_data('Manhattan',"2025-07-01","2025-08-01",'hour',['SRAVG'])
df.head(3)
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, figsize=(6, 4))
charts.plot_temperature(axes[0], df, ["TEMP2MAVG","TEMP2MMAX","TEMP2MMIN"])
charts.plot_solar_radiation(axes[1], df, "SR")
plt.show()
```

### Get soil moisture

```python
# Download daily temperature and precipitation for one station
df = ms.request_data(
    station="Manhattan",
    start="2024-01-01",
    end="2025-01-01",
    interval="day",
    variables=["SOILKA5CM","SOILKA10CM","SOILKA20CM","SOILKA50CM",
             "SOILEC5CM","SOILEC10CM","SOILEC20CM","SOILEC50CM"],
)

df.head(3)
```

```python
df = ms.calibrate_vwc(df)
df.head(3)
```

```python
df = ms.compute_soil_water_storage(df)
df.head(3)
```

```python
fig, ax = plt.subplots(1, 1, sharex=True, figsize=(6, 4))
charts.plot_vwc(ax=ax, df=df, variables=['VWC5CM','VWC10CM','VWC20CM','VWC50CM'])
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(10, 8))
ms.plot_temperature(axes[0], df, ["TEMP2MAVG", "TEMP2MMIN", "TEMP2MMAX"])
ms.plot_precip(axes[1], df, "PRECIP")
ms.plot_humidity(axes[2], df, "RELHUM2MAVG")
```
